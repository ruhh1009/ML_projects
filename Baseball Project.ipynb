# Baseball Project

This dataset utilizes data from 2014 Major League Baseball seasons in order to develop an algorithm that predicts the number of wins for a given team in the 2015 season based on several different indicators of success. There are 16 different features that will be used as the inputs to the machine learning and the output will be a value that represents the number of wins. 

1 Input features: Runs, At Bats, Hits, Doubles, Triples, Homeruns, Walks, Strikeouts, Stolen Bases, Runs Allowed, Earned Runs, Earned Run Average (ERA), Shutouts, Saves, and Errors

2 Output: Number of predicted wins (W)

```python
# Declaration of independent variable.
x=df.drop('W', axis=1)
```

```python
# Declaration of target variable.
y=df['W']
```

```python
x.skew()
```

```python
from sklearn.preprocessing import power_transform
pt=power_transform(x, method='yeo-johnson')
x=pd.DataFrame(pt, columns=x.columns)
x.skew()
```

```python
x
```

```python
x=x.drop(['R','AB','H'], axis=1)
```

```python
x
```

```python
# Checking outlier.
x.boxplot(figsize=[20,8])
```

# Model building

```python
from sklearn.model_selection import train_test_split
```

```python
# splitting into train and validation with 30% data in validation set and 70% data in train set.
x_train, x_test, y_train, y_test=train_test_split(x,y,random_state=5, test_size=0.3)
```

Now our data is ready. Its time to build our model and check its performance. Linear regression is used for continous problems and as it is a continous problem let's first build a Linear Regression model.

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()       # defining the logistic regression model
lr.fit(x_train,y_train)     # fitting the model on  X_train and y_train
predlr=lr.predict(x_test)   # making prediction on the validation set
```

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

Now we will evaluate how accurate our predictions are. As the evaluation metric for this problem is mean squared error and root mean squared error, let's calculate the that on validation set.

```python
# Evaluting the metrics
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test,predlr))
print('Mean Squared Error:', metrics.mean_squared_error(y_test,predlr))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test,predlr)))
```

```python
from sklearn.model_selection import cross_val_score
scr1=cross_val_score(lr,x,y,cv=6)
print('Cross Validation Score is :', scr1.mean())
```

# Lasso Analysis

```python
from sklearn.linear_model import Lasso
reg1=Lasso(alpha=50,max_iter=100, tol=0.1)
reg1.fit(x_train,y_train)
reg1.score(x_train,y_train)
```

```python
reg1.score(x_test,y_test)
```

# Ridge Analysis

```python
from sklearn.linear_model import Ridge
reg2=Ridge(alpha=50,max_iter=100, tol=0.1)
reg2.fit(x_train,y_train)
reg2.score(x_train,y_train)
```

```python
reg2.score(x_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
scr2=cross_val_score(reg2,x,y,cv=6)
print('Cross Validation Score is :', scr2.mean())
```

Minimum difference between train data and test data scores is for Ridge(0.11), so this is our best model.
