Question 2: Scrape the following details of each product listed in first 3 pages of your search results and save it in a dataframe and csv. In case if any product vertical has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
#creating the empty list
url=[]
brand_name=[]
product_name=[]
details=[]
star=[]
rating_number=[]
exchange=[]
price=[]
delivery=[]
available=[]


#scrapping the required details
start=0
end=2
for page in range(start,end):#for loop for scrapping 3 page
    for i in driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']"):
        url.append(i.get_attribute("href"))


time.sleep(3)
for i in url:
    driver.get(i)
    time.sleep(3)
    #for scrapping the brand name
    try:
        brand=driver.find_element_by_id("bylineInfo")
        brand_name.append(brand.text)
    except NoSuchElementException:
        brand_name.append('-')
    
    #for scrapping the product name
    try:
        name=driver.find_element_by_id("productTitle")
        product_name.append(name.text)
    except NoSuchElementException:
        product_name.append('-')
    
    #for scrapping the other details of product
    try:
        desc=driver.find_element_by_id("detailBullets_feature_div")
        details.append(desc.text.replace('\n',''))
    except NoSuchElementException:
        details.append('-')
    
    #for scrapping the ratings of product
    try:
        rating=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
        star.append(rating.text)
    except NoSuchElementException:
        star.append('-')  
    
    #for scrapping the number of ratings
    try:
        number=driver.find_element_by_xpath("//span[@class='a-size-base a-color-secondary']")
        rating_number.append(number.text)
    except NoSuchElementException:
        rating_number.append('-')
    
    #for scrapping the return/exchange
    try:
        save=driver.find_element_by_xpath("//td[@class='a-span12 a-color-price a-size-base priceBlockSavingsString']")
        exchange.append(save.text)
    except NoSuchElementException:
        exchange.append('-')
    
    #for scrapping the price of product
    try:
        cost=driver.find_element_by_id("priceblock_ourprice")
        price.append(cost.text)
    except NoSuchElementException:
        price.append('-')
    
    #for scrapping the expected delivery
    try:
        edelivery=driver.find_element_by_id("ddmDeliveryMessage")
        delivery.append(edelivery.text)
    except NoSuchElementException:
        delivery.append('-')
    
    #for scrapping the availability
    try:
        avail=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-success']")
        available.append(avail.text)
    except NoSuchElementException:
        available.append('-')
        

time.sleep(4)
# creating the dataframe from the scraped data
df=pd.DataFrame({"brand_name":brand_name,"product_name":product_name,"product_url":url,"star":star,"delivery":delivery,
                 "rating_number":rating_number,"price":price,"exchange":exchange,"details":details,"available":available})
df
```

Question 4: To search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Display Resolution”, “Processor”, “Processor Cores”, “Battery Capacity”, “Price”, “Product URL”.
```python
# Opening the homepage of flipkart
driver.get('https://www.flipkart.com')
time.sleep(10)

# Locating search bar
search4job= driver.find_element_by_xpath("//input[@class='_3704LK']")

# Writing on search bar
search4job.send_keys("smartphone")


# do click using xpath function
search4btn= driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
search4btn.click()
```

Question 6: Scrap details of all the funding deals for second quarter (i.e. July 20 – September 20) from trak.in.

```python
# Opening the page of funding deals from trak.in
driver.get('https://trak.in/india-startup-funding-investment-2015/')
```

```python
#creating the empty list
Date=[]
Startup_Name=[]
Industry=[]
Sub_Vertical=[]
Location=[]
Investor=[]
Investment_Type=[]
Amount=[]

# scraping the required data
for i in range(48,51):

    driver.find_element_by_xpath('//div[@id="tablepress-{}_wrapper"]/div/label/select/option[4]'.format(i)).click()



    # for scraping the Date

    dt = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[2]'.format(i))

    for d in dt:

        Date.append(d.text)



    # for scraping the Startup Name

    sn = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[3]'.format(i))

    for n in sn:

        Startup_Name.append(n.text)

    

    # for scraping the Industry/Vertical

    ind = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[4]'.format(i))

    for n in ind:

        Industry.append(n.text)

    

    # for scraping the Sub-Vertical

    sv = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[5]'.format(i))

    for s in sv:

        Sub_Vertical.append(s.text)



    # for scraping the Location

    loc = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[6]'.format(i))

    for l in loc:

        Location.append(l.text)

    

    # for scraping the Investor

    inv = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[7]'.format(i))

    for n in inv:

        Investor.append(n.text)

    

    # for scraping the Investment Type

    invt = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[8]'.format(i))

    for n in invt:

        Investment_Type.append(n.text)

    

    # for scraping the Amount

    amt = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[9]'.format(i))

    for a in amt:

        Amount.append(a.text)

    
time.sleep(4)
# creating the dataframe from the scraped data
fund_df = pd.DataFrame({"Date":Date,"Startup_Name":Startup_Name,"Industry":Industry,"Sub_Vertical":Sub_Vertical,
                       "Location":Location,"Investor":Investor,"Investment_Type":Investment_Type,"Amount":Amount})

fund_df

```

Question 7: Scrap all the available details of top 10 gaming laptops from digit.in.

```python
# Opening the page of top 10 list of best budget gameing laptop in digit.in website
driver.get('https://www.digit.in/top-products/best-budget-gaming-laptop-3591.html')
time.sleep(10)
```

```python
#creating the empty list
name=[]
desc=[]
OS=[]
display=[]
processor=[]
memory=[]
weight=[]
dimension=[]
G_processor=[]
price=[]
urls=[]


#scraping the description
details=driver.find_elements_by_xpath("//div[@class='Section-center']/p[1]")
for i in details:
    if i.text is None :
        desc.append("--") 
    else:
        desc.append(i.text)
        
time.sleep(4)        
#scraping the operating system 
system=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[3]/td[3]")
for i in system:
    if i.text is None :
        OS.append("--") 
    else:
        OS.append(i.text)


time.sleep(4)
#scraping the display 
screen=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[4]/td[3]")
for i in screen:
    if i.text is None :
        display.append("--") 
    else:
        display.append(i.text)

        
time.sleep(4)
#scraping the processor
cpu=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[5]/td[3]")
for i in cpu:
    if i.text is None :
        processor.append("--") 
    else:
        processor.append(i.text)
        
        
time.sleep(4)
#scraping the memory
space=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[6]/td[3]")
for i in space:
    if i.text is None :
        memory.append("--") 
    else:
        memory.append(i.text)
        
        
time.sleep(4)
#scraping the weight 
laptop_weight=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[7]/td[3]")
for i in laptop_weight:
    if i.text is None :
        weight.append("--") 
    else:
        weight.append(i.text)
        
        
time.sleep(4)
#scraping the dimension
size=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[8]/td[3]")
for i in size:
    if i.text is None :
        dimension.append("--") 
    else:
        dimension.append(i.text)
        
        
        
time.sleep(4)
#scraping the G_processor
gprocessor=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[9]/td[3]")
for i in gprocessor:
    if i.text is None :
        G_processor.append("--") 
    else:
        G_processor.append(i.text)
        
# scraping of urls from xpath
for i in driver.find_elements_by_xpath("//div[@class='right-container']/div/a"):
    urls.append(i.get_attribute("href"))
    
# scraping of laptop name and price from urls
for i in urls:
    driver.get(i)
    try:
        laptop_name=driver.find_element_by_xpath("//div[@class='heading-wraper']/h1")
        name.append(laptop_name.text)
    except NoSuchElementException:
        name.append('-')
    
    
    try:
        laptop_price=driver.find_element_by_xpath("//div[@class='Block-price']/b")
        price.append(laptop_price.text)
    except NoSuchElementException:
        price.append('-')

time.sleep(4)
# creating the dataframe from the scraped data        
df=pd.DataFrame({"laptop_name":name,"description":desc,"operation_syatem":OS,
                "display":display,"processor":processor,"memory":memory,"weight":weight,
                 "dimension":dimension,"Graphic_processor":G_processor,"price":price})
df
```
