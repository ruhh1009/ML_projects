```python
# Import all the required libraries
import pandas as pd
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import requests
```

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe")
```

```python
# Opening the homepage of flipkart
driver.get('https://www.flipkart.com')
time.sleep(10)

# Locating search bar
search4job= driver.find_element_by_xpath("//input[@class='_3704LK']")

# Writing on search bar
search4job.send_keys("sarees")


# do click using xpath function
search4btn= driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
search4btn.click()
```

```python
img_urls = []
img_data = []
time.sleep(2)
#scrapping the required details
start=0
end=14
for page in range(start,end):
    # for collecting the urls
    time.sleep(30)
    # Image urls xpath
    images = driver.find_elements_by_xpath('//img[@class="_2r_T1I"]')
    for image in images:
        source= image.get_attribute('src')
        if source is not None:
            if(source[:4] == 'http'):
                img_urls.append(source)
    img_urls=img_urls[:500]
                    
    nxt_button=driver.find_elements_by_xpath("//a[@class='_1LKTO3']")#scraping the list of buttons from the page
    try:
        driver.get(nxt_button[1].get_attribute('href'))#getting the link from the list for next page
    except:
        driver.get(nxt_button[0].get_attribute('href'))

for i in range(len(img_urls)):
    if i >= 500:
        break
    print("Downloading {0} of {1} images" .format(i, 500))
    response= requests.get(img_urls[i])
    file = open(r"C:\Users\PRIYA\Pictures\Sarees(women)\sarees"+str(i)+".jpg", "wb")
    file.write(response.content)
```
