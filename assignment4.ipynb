Question 1: Scrape the details of most viewed videos on YouTube from Wikipedia:

Url= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos

You need to find following details:

A) Rank

B) Name

C) Artist

D) Upload date

E) Views

```python
# Opening the required url as given in question.
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
#creating the empty list
Rank=[]
Name=[]
Artist=[]
Upload_date=[]
Views=[]

#scraping name from xpath
nvedio=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')
for i in nvedio:
    Name.append(i.text)         #appending the text in name
Name=Name[:30]

#scraping rank from xpath
rvedio=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
for i in rvedio:
    Rank.append(i.text)         #appending the text in rank
Rank=Rank[:30]

#scraping Artist from xpath
avedio=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')
for i in avedio:
    Artist.append(i.text)       #appending the text in artist
Artist=Artist[:30]

#scraping Upload date from xpath
uvedio=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
for i in uvedio:
    Upload_date.append(i.text)  #appending the text in Upload date
Upload_date=Upload_date[:30]

#scraping Views from xpath
vvedio=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
for i in vvedio:
    Views.append(i.text)        #appending the text in Views
Views=Views[:30]


time.sleep(4)
# creating the dataframe from the scraped data
df1 = pd.DataFrame({"Rank":Rank,"Name":Name,"Artist":Artist,"Views(billions)":Views,"Upload_date":Upload_date})

df1
```

Question 5: Scrape the details of trending repositories on Github.com.

Url = https://github.com/

You have to find the following details:

A) Repository title

B) Repository description

C) Contributors count

D) Language used

Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
# Opening the homepage of github.com
driver.get('https://github.com/')

time.sleep(5)
# do click on explore
driver.find_element_by_xpath('//nav[@class="mt-0 px-3 px-lg-0 mb-5 mb-lg-0"]/ul/li[4]/details').click()
time.sleep(5)
# do click on trending
driver.find_element_by_xpath('//div[@class="dropdown-menu flex-auto rounded px-0 pt-2 pb-0 mt-0 pb-4 p-lg-4 position-relative position-lg-absolute left-0 left-lg-n4"]/ul[2]/li[3]').click()
```

```python
#creating the empty list
re_title=[]
r_description=[]
contributor=[]
language=[]
url=[]

#scraping title from xpath
title=driver.find_elements_by_xpath('//div[@class="position-relative container-lg p-responsive pt-6"]/div/div[2]/article/h1')
for i in title:
    re_title.append(i.text)     #appending the text in title
     

# for scraping the repositories url
for i in driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a"):
    url.append(i.get_attribute("href"))
        

time.sleep(3)
# scrapping the required data from url
for i in url:
    driver.get(i)
    time.sleep(3)
    #for scrapping the description
    try:
        desc=driver.find_element_by_xpath("//p[@class='f4 mt-3']")
        r_description.append(desc.text)
    except NoSuchElementException:
        r_description.append('-')
        
    #for scrapping the contributors count
    try:
        count=driver.find_element_by_xpath("//h2[@class='h4 mb-3']/a[contains(text(),'Contributors')]/span")
        contributor.append(count.text)
    except:#handling no such element exception
        contributor.append('No details available')
        

    #for scrapping the language used
    l=[]
    languages=driver.find_elements_by_xpath("//li[@class='d-inline']")
    if languages:   
        for i in languages:
            a=str(i.text)
            b=a.split("\n",)
            l.append(b[0])
    else:
        l.append('No languages used')
    language.append(l)
    
time.sleep(4)
# creating the dataframe from the scraped data
df5 = pd.DataFrame({"re_title":re_title,"r_description":r_description,"contributor":contributor,
                   "language":language})

df5
```

```python
#creating the empty list
name=[]
d_type=[]
Task=[]
Attribute_type=[]
instances=[]
No_attribute=[]
Year=[]
url=[]

#scraping votes from xpath
Name=driver.find_elements_by_xpath("//p[@class='normal']/b/a")
for i in Name:
    name.append(i.text)      #appending the text in votes

#scraping votes from xpath
year=driver.find_elements_by_xpath("//table[@cellpadding='5']/tbody//td[7]")
for i in year:
    Year.append(i.text)      #appending the text in votes
Year=Year[1:589:1]
Year

    
# for scraping url of all machine learning program list
for i in driver.find_elements_by_xpath("//p[@class='normal']/b/a"):
    url.append(i.get_attribute("href"))
        

time.sleep(3)
# scraping of all required data from the urls
for i in url:
    driver.get(i)
    time.sleep(3)
    #for scrapping the data type
    try:
        types=driver.find_element_by_xpath("//table[@cellpadding='6']/tbody/tr[1]/td[2]/p")
        d_type.append(types.text)
    except NoSuchElementException:
        d_type.append('-') 
        
    #for scrapping the number of instance
    try:
        inst=driver.find_element_by_xpath("//table[@cellpadding='6']/tbody/tr[1]/td[4]/p")
        instances.append(inst.text)
    except NoSuchElementException:
        instances.append('-')
        
    #for scrapping the attribute type
    try:
        Attribute=driver.find_element_by_xpath("//table[@cellpadding='6']/tbody/tr[2]/td[2]/p")
        Attribute_type.append(Attribute.text)
    except NoSuchElementException:
        Attribute_type.append('-')
        
    #for scrapping the number of attribute
    try:
        number_att=driver.find_element_by_xpath("//table[@cellpadding='6']/tbody/tr[2]/td[4]/p")
        No_attribute.append(number_att.text)
    except NoSuchElementException:
        No_attribute.append('-')
        
    #for scrapping the Task
    try:
        task=driver.find_element_by_xpath("//table[@cellpadding='6']/tbody/tr[3]/td[2]/p")
        Task.append(task.text)
    except NoSuchElementException:
        Task.append('-') 
        

time.sleep(4)
# creating the dataframe from the scraped data
df10 = pd.DataFrame({"Dataset_name":name,"Year":Year,"datatype":d_type,"Task":Task,"Attributes_type":Attribute_type,
                    "No_of_instances":instances,"No_of_attributes":No_attribute})

df10
```
